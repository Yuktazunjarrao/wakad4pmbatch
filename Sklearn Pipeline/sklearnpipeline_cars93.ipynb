## Using Automatic Sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
# Numeric Pipeline
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
# Categorical Pipeline
cat_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                           ('ohe',OneHotEncoder())])
# Combine numeric and categorical pipeline
pre =ColumnTransformer([('num',num_pipe,con),
                        ('cat',cat_pipe,cat)])
```

```python
X_pre = pre.fit_transform(X).toarray()
X_pre
```

```python
cols = pre.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

```python
X_pre.isna().sum()
```

```python
# Save the preproessed dataset
X_pre.to_csv('carspre1.csv',index=False)
```

## Train and Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest =train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Creating a model

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
model.fit(xtrain, ytrain)
```

```python
#R2 score for training
model.score(xtrain, ytrain)
```

```python
#R2 score for testing
model.score(xtest, ytest)
```

## Predict Train and Test data 

```python
ypred_tr = model.predict(xtrain)
ypred_ts = model.predict(xtest) 
```

## Evaluating Train and Test data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def evaluate_model(xtrain, ytrain, xtest, ytest, model):
    ypred_tr = model.predict(xtrain)
    ypred_ts = model.predict(xtest)
    
    # Trainiing results
    tr_mse = mean_squared_error(ytrain, ypred_tr)
    tr_rmse = tr_mse**(1/2)
    tr_mae = mean_absolute_error(ytrain,ypred_tr)
    tr_r2 = r2_score(ytrain,ypred_tr)
    print('Training Results :\n')
    print(f'MSE : {tr_mse:.2f}')
    print(f'RMSE: {tr_rmse:.2f}')
    print(f'MAE : {tr_mae:.2f}')
    print(f'R2  : {tr_r2:.4f}')
    print('\n========================================\n')

    # Testing result
    ts_mse = mean_squared_error(ytest, ypred_ts)
    ts_rmse = ts_mse**(1/2)
    ts_mae = mean_absolute_error(ytest,ypred_ts)
    ts_r2 = r2_score(ytest,ypred_ts)
    print('Testing Results :\n')
    print(f'MSE : {ts_mse:.2f}')
    print(f'RMSE: {ts_rmse:.2f}')
    print(f'MAE : {ts_mae:.2f}')
    print(f'R2  : {ts_r2:.4f}')
```

```python
evaluate_model(xtrain, ytrain, xtest, ytest, model)
```
